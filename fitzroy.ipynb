## 압구정 슈프림

```python
apg_sup = pd.read_excel('fitzroy.xlsx', usecols = 'A:S')
# apg_pal = pd.read_excel('fitzroy.xlsx', sheet_name = 1)
# apg_acc = pd.read_excel('fitzroy.xlsx', sheet_name = 2)
# hyundai =  pd.read_excel('fitzroy.xlsx', sheet_name = 3)
```

```python
apg_sup.head()
```

```python
apg_sup.columns
```

```python
# keep unique row records 
sheet1 = apg_sup.loc[:'146',['품명', '이름', '색상',' S .2',
                  'M.2', 'L.2', 'XL.2', '현재 재고']]
sheet1 = sheet1.drop_duplicates(subset=['품명', '이름', '색상'])
```

```python
sheet1.head()
```

```python
sheet1 = sheet1.rename(columns={" S .2": "S", 'M.2':'M', 
                                'L.2':'L', 'XL.2':'XL'})
sheet1.head()
```

```python
# melt

sheet1_melted = sheet1.melt(id_vars = ('품명','이름','색상', '현재 재고'), 
                            var_name = '사이즈', value_name = '재고')
```

```python
sheet2 = apg_sup.loc[158:,['품명', '이름', '색상', ' S ', 'M', 
                           'L', 'XL', '20.08.22 사이즈 재고',
                           ' S .1', 'M.1', 'L.1', 'XL.1', 'Unnamed: 13']]
sheet2 = sheet2.drop_duplicates(subset=['품명', '이름', '색상'])
```

```python
# sheet2 = sheet2.rename(columns={' S ': "S", 
#                                 '20.08.22 사이즈 재고':"현재 재고"})
# sheet2.head()

# sheet2 = sheet2.fillna(0)
 
sheet2 = sheet2.assign(
    S = lambda x: x[' S '] - x[' S .1']
).assign(
    M = lambda x: x['M'] - x['M.1']
).assign(
    L = lambda x: x['L'] - x['L.1']
).assign(
    XL = lambda x: x['XL'] - x['XL.1']
)
```

```python
sheet2 = sheet2[['품명','이름','색상','S','M','L','XL', 'Unnamed: 13']]
sheet2.head()
```

```python
sheet2 = sheet2.rename(columns={'Unnamed: 13': "현재 재고"})
```

```python
# melt

sheet2_melted = sheet2.melt(id_vars = ('품명','이름','색상', '현재 재고'), 
                            var_name = '사이즈', value_name = '재고')
```

```python
# save
# sheet1.to_excel("sup_apg.xlsx", sheet_name = "sheet1")

with pd.ExcelWriter('fitzroy_new.xlsx') as writer:  
    sheet1_melted.to_excel(writer, sheet_name='Sheet1')
    sheet2_melted.to_excel(writer, sheet_name='Sheet2')
```

## 압구정 슈프림 매출

```python
sup_sales = pd.read_excel('fitzroy.xlsx', usecols = 'A:C,I:M')
```

```python
sup_sales.head()
```

```python
sup_sales = sup_sales.fillna(0)
```

```python
sup_sales_split = sup_sales['판매날짜&사이즈'].str.split(",").apply(pd.Series)
sup_sales_split.index = sup_sales.set_index(['품명','이름','색상']).index
sup_sales_split = sup_sales_split.stack().reset_index(['품명','이름','색상'])
sup_sales_split.columns = sup_sales_split.columns.map(str)
sup_sales_split = sup_sales_split.reset_index()
sup_sales_split = sup_sales_split.rename(columns={"0":"판매"})
```

```python
sup_sales_split['날짜'], sup_sales_split['사이즈'] = sup_sales_split['판매'].str.split('(', 1).str
```

```python
sup_sales_split['사이즈'] = sup_sales_split['사이즈'].str.rstrip(')')
```

```python
sup_sales_split
```

```python
with pd.ExcelWriter('fitzroy_new.xlsx', mode='a') as writer:  
    sup_sales_split.to_excel(writer, sheet_name='압구정매출')
```

## 압구정 팰리스

```python
apg_pal = pd.read_excel('fitzroy.xlsx', usecols = 'B:H,J:M,O:R,T:Y',
                        sheet_name = 1)
```

```python
apg_pal.head()
```

```python
apg_pal = apg_pal.fillna(0)
apg_pal.columns
```

```python
# keep unique row records 
sheet3 = apg_pal.loc[:'25',['품목', '이름', '색상',
                             'M.2', 'L.2', 'XL.2', '현재 총 재고',]]
sheet3 = sheet3.drop_duplicates(subset=['품목', '이름', '색상'])
sheet3 = sheet3.rename(columns = {'M.2':'M', 'L.2':'L', 'XL.2':'XL'})
```

```python
sheet4 = apg_pal.loc['35':,['품목', '이름', '색상',
                            'M', 'L', 'XL', '20.08.22 사이즈 재고',
                            'M.1', 'L.1', 'XL.1', '현재 재고']]
sheet4 = sheet4.drop_duplicates(subset=['품목', '이름', '색상'])

sheet4 = sheet4.assign(
    M = lambda x: x['M'] - x['M.1'],
    L = lambda x: x['L'] - x['L.1'],
    XL = lambda x: x['XL'] - x['XL.1'],
)

sheet4 = sheet4[['품목', '이름', '색상', 'M', 'L', 'XL', '현재 재고']]
```

```python
sheet4
```

```python
# melt

sheet3_melted = sheet3.melt(id_vars = ('품목','이름','색상', '현재 총 재고'), 
                            var_name = '사이즈', value_name = '재고')
sheet4_melted = sheet4.melt(id_vars = ('품목','이름','색상', '현재 재고'), 
                            var_name = '사이즈', value_name = '재고')
```

```python
sheet4_melted
```

```python
# save
# sheet1.to_excel("sup_apg.xlsx", sheet_name = "sheet1")

with pd.ExcelWriter('fitzroy_new.xlsx') as writer:  
    sheet3_melted.to_excel(writer, sheet_name='Sheet3')
    sheet4_melted.to_excel(writer, sheet_name='Sheet4')
```
